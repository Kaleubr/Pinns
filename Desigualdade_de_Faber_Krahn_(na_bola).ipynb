$\textbf{Desigualdade de Faber-Krahn}$

Quando $\Omega\subset\mathbb{R}^{N}$ é uma bola, podemos resolver de maneira explicita para o autovalor principal e sua autofunção o caso do operador laplaciano, utilizando funções de Bessel.

Considere o seguinte problema de Dirchlet:

\begin{align}
 \begin{cases}
\Delta u = \lambda u,\hbox{  }u\in\Omega\subset\mathbb{R}^{N}\\
\hbox{  }\hbox{  }\hbox{  }u \equiv 0,\quad u\in\partial \Omega
 \end{cases}
\end{align}

Seja $\Omega = \Omega^{*} = B(0,R)$ a bola de centro na origem e de raio $R$. Como a simetrização de Schwarz decresce a integral

$$ \int_{\Omega}|\varphi_{1}|^{2}\hbox{  }dx$$

e não altera a norma de $L^{2}(\Omega)$, segue que

$$ R_{A}(\varphi_{1})\ge R_{A}(\varphi_{1}^{*}). $$

onde $R_{A}(u) = \frac{a(u,u)}{\|u\|^{2}_{2,\Omega}} $ é o quociente de Rayleigh e $a(u,v)$ uma forma bilinear definida em $H_{0}^{1}(\Omega)\times H_{0}^{1}(\Omega)$ descrita por

$$ a(u,v) = \int_{\Omega}\nabla u\cdot\nabla v\hbox{  }dx. $$

Como $\lambda_{1}(\Omega)$ é o mínimo do quociente de Rayleigh, segue que $\varphi_{1}^{*}$ atinge seu mínimo, e consequentemente é autofunção de $\lambda_{1}(\Omega)$. Segue então da simplicidade do autovalor principal que

$$ \varphi_{1} = \varphi_{1}^{*}. $$

Portanto a primeira autofunção do laplaciano para o problema de Dirchlet é radial e decrescente.

Faça $\varphi_{1}(x) = v(|x|)$. Vamos descrever $v$ através de seu raio (Isso é possível na bola pelo Lema de Gidas). De fato, note que

$$ |x| = r\hbox{  } ⇒ x = r^{2},$$

derivando temos que

\begin{align*}
    &v_{x_{i}} = \frac{\partial}{\partial x_{i}}[v(r)] = \frac{x_{i}v'(r)}{r}\\
    &v_{x_{i}x_{i}} = \frac{\frac{\partial}{\partial x_{i}}[x_{i}v'(r)]r -(x_{i}v'(r))r_{x_{i}}}{r^{2}}\\
    &v_{x_{i}x_{i}} = \frac{\left[v'(r) + x_{i}\left(v''(r)\frac{x_{i}}{r}\right)\right]r - \frac{x_{i}v'(r)x_{i}}{r}}{r^{2}}\\
    &v_{x_{i}x_{i}} = \frac{rv'(r) + x_{i}v''(r) - \frac{v'(r)x^{2}_{i}}{r}}{r^{2}}.
\end{align*}

Logo temos que
\begin{align*}
    \Delta v &= \sum_{i = 1}^{N}v_{x_{i}x_{i}}\\
             &=\frac{1}{r^{2}}\left( r\sum_{i = 1}^{N}v'(r) + v''(r)\sum_{i = 1}^{N}x^{2}_{i} - \frac{v'(r)}{r}\sum_{i = 1}^{N}x^{2}_{i} \right)\\
             &=\frac{1}{r^{2}}\left( rNv'(r) + r^{2}v''(r) - rv'(r) \right)\\
             &=\frac{Nv'(r)}{r}+v''(r) - rv'(r)\\
             &= -v''(r) - \left(\frac{N-1}{r}\right)v'(r).
\end{align*}

Se considerarmos o problema de autovalor, temos a seguinte equação:

$$ v''(r) + \left(\frac{N-1}{r}\right)v'(r) + \lambda v(r) = 0, $$

onde $v'(R) = v(0) = 0$. Iremos mostrar que

$$ v(r) = cr^{1 - \frac{N}{2}}J_{\frac{N}{2}-1}(\sqrt{\lambda}r) $$

é solução do problema de autovalor, com $c$ sendo uma constante arbitrária e $J_{p}$ sendo a função de Bessel (de tipo um) ordem $p$ que satisfaz a equação diferencial

$$ \frac{d^{2}y}{d\rho^{2}}(\rho) + \frac{1}{\rho}\frac{dy}{d\rho}(\rho) +\left( 1 - \frac{p^{2}}{\rho^{2}}\right) y(\rho) = 0. $$

De fato, faça $v(r) = r^{\nu}\omega(r)$. Calculando as derivadas temos que

$$v'(r) = \nu r^{\nu - 1}\omega(r) + r^{\nu}\omega'(r),$$

$$ v''(r) = \nu(\nu - 1)r^{\nu -2}\omega(r) + 2\nu^{\nu-1} + r^{\nu}\omega''(r) $$

Substituindo na EDO temos

$$ [\nu(\nu - 1)r^{\nu -2}\omega(r) + 2\nu^{\nu-1} + r^{\nu}\omega''(r)] + \left( \frac{N-1}{r} \right)[\nu r^{\nu - 1}\omega(r) + r^{\nu}\omega'(r)] + \lambda r^{\nu}\omega(r) = 0 $$

Agrupando termos semelhantes

$$[\nu^{2} + (N-2)\nu]r^{\nu-2}\omega(r) + [2\nu + (N-1)]r^{\nu-1}\omega'(r) + [\omega''(r) + \lambda(r)]r^{\nu} = 0. $$

Faça $\nu = 1 - \frac{N}{2}$. Temos que

$$ \left[\left(1 - \frac{N}{2}\right)^{2} + (N-2)\left(1 - \frac{N}{2}\right)\right]r^{\nu-2}\omega(r) + \left[ 2\left(1 - \frac{N}{2}\right) + (N-1) \right]r^{\nu-1} + [\omega''(r) + \lambda\omega(r)]r^{\nu} = 0 $$

Note que os termos multiplicados por $r^{\nu-2}$ se anulam e os termos de $r^{\nu -1}$ são iguais a 1. Logo temos

$$ r^{\nu}\omega''(r) + r^{\nu -1}\omega'(r) + r^{\nu}\lambda\omega(r) = 0$$

Dividndo ambos os lados por $r^{\nu}$

$$ \omega''(r) + \frac{1}{r}\omega'(r) + \lambda\omega(r) = 0 $$

Comparando com a equação de Bessel modificada temos que

$$ \omega''(r) + \frac{1}{r}\omega'(r) + \left(\lambda - \frac{\left(\frac{N}{2} - 1\right)^{2}}{r^{2}}\right)\omega(r) = 0 $$

Faça $z = \sqrt{\lambda}r$, então

$$ \frac{d}{dr} = \sqrt{\lambda}\frac{d}{dz} $$

$$ \frac{d^{2}}{dr^{2}} = \lambda\frac{d^{2}}{dz^{2}} $$

substituindo na equação anterior temos que

$$ \lambda \frac{d^{2}\omega}{dz^{2}} + \frac{\sqrt{\lambda}}{(z/\sqrt{\lambda})}\frac{d\omega}{dz} +  \left( \lambda - \frac{\left( \frac{N}{2} - 1 \right)^{2}}{(z^{2}/\lambda)} \right)\omega(z) = 0$$

dividindo tudo por $\lambda$ temos

$$ \frac{d^{2}\omega}{dz^{2}} + \frac{1}{z}\frac{d\omega}{dz} + \left( 1 - \frac{\nu^{2}}{z^{2}}\right)\omega(z) = 0 $$

onde $\nu = \frac{N}{2} - 1$. A solução geral da equação de Bessel é da forma

$$ \omega(z) = c_{1}J_{\nu}(z) + c_{2}Y_{\nu}(z) $$

Onde $J_{\nu}$ é a função de Bessel de primeira espécie e $Y_{\nu}$ é a função de Bessel de segunda espécie. Para nossa aplicação iremos considerar somente $J_{\nu}$ pois quando $r\to 0$ temos que $Y_{\nu}\to \infty$ quando $\nu$ é não-inteiro. Logo

$$ \omega(z) = cJ_{\nu}(z). $$

Como $z = \sqrt{\lambda}r$,

$$ \omega(r) = cJ_{\frac{N}{2} - 1}(\sqrt{\lambda}r) $$

Como $\omega(r) = r^{\nu}v(r)$ concluímos que

$$ v(r) = cr^{1 - \frac{N}{2}}J_{\frac{N}{2} - 1}(\sqrt{\lambda}r). $$


Precisamos mostrar que o autovalor principal do operador laplaciano na bola $B_{R} = B(0,R)$ é da forma

$$ \lambda_{1}(B_{R}) = \frac{j^{2}_{\frac{N}{2} - 1, 1}}{R^{2}} $$

De fato, como $v(R) = 0$, temos que

$$ v(R) = cR^{1 - \frac{N}{2}}J_{\nu - 1}(\sqrt{\lambda}R) = 0 $$

isso implica que

$$ J_{\frac{N}{2}-1}(\sqrt{\lambda}R) = 0. $$

Denote por $j_{\frac{N}{2} - 1,1}$ o primeiro zero positivo de $J_{\frac{N}{2} -1}$. Logo

$$ \sqrt{\lambda_{1}}R = j_{\frac{N}{2} - 1, 1}$$

portanto

$$ \lambda_{1}(B_{R}) = \frac{j^{2}_{\frac{N}{2} - 1, 1}}{R^{2}}$$

Precisamos mostrar que a autofunção com respeito a ao autovalor principal do operador laplaciano é da forma

$$ \varphi_{1}(x) = |x|^{1 - \frac{N}{2}}J_{\frac{N}{2} - 1}\left( \frac{j_{\frac{N}{2} - 1,1}}{R}|x| \right) $$

Como $\varphi_{1}(x) = v(|x|)$, concluímos que

$$ \varphi_{1}(x) = c|x|^{1 - \frac{N}{2}}J_{\frac{N}{2} - 1}\left( \frac{j_{\frac{N}{2} - 1,1}}{R}|x| \right) $$

onde $c$ é uma constante arbitrária. Vamos normalizar a função $\varphi_{1}$ afim de encontrarmos o valor de $c$. De fato, como vamos normalizar em $L^{2}(\Omega)$, então $\varphi_{1}$ satisfaz

$$ \int_{\Omega}|\varphi_{1}(x)|\hbox{  }dx = 1. $$

Como $\varphi_{1}(x)$ é radial, temos que passar para coordenadas esféricas, isto é,

$$ dx = r^{N-1}drd\sigma $$

onde $d\sigma$ é o elemento de área da esféra unitária $\mathbb{S}^{N - 1}$. Como $\Omega = B(0,R)$, temos que

$$ \int_{\Omega}|\varphi_{1}(x)|^{2}\hbox{  }dx = \int_{0}^{R}|\varphi_{1}(r)|^{2}r^{N-1}\hbox{  }dr\int_{\mathbb{S}^{N-1}}d\sigma. $$

Como $|\varphi_{1}(r)|^{2} = r^{2 - N}J_{\frac{N}{2} - 1}^{2}\left(\dfrac{j_{\frac{N}{2} - 1,1}}{R}r\right)$, temos que

\begin{align*}
\int_{\Omega}|\varphi_{1}(x)|^{2}\hbox{  }dx &= \frac{2\pi^{\frac{N}{2}}}{\Gamma\left(\frac{N}{2}\right)}\int_{0}^{R}r^{2 - N}J_{\frac{N}{2} - 1}^{2}\left(\frac{j_{\frac{N}{2} - 1,1}}{R}r\right)r^{N-1}\hbox{  }dr\\
&=\frac{2\pi^{\frac{N}{2}}}{\Gamma\left(\frac{N}{2}\right)}\int_{0}^{R}rJ_{\frac{N}{2} - 1}^{2}\left(\frac{j_{\frac{N}{2} - 1,1}}{R}r\right)\hbox{  }dr,
\end{align*}

faça $s = \left(\dfrac{j_{\frac{N}{2} - 1,1}}{R}r\right)$, então $ds = \left(\dfrac{j_{\frac{N}{2} - 1,1}}{R}\right)dr$. Logo

\begin{align*}
\frac{2\pi^{\frac{N}{2}}}{\Gamma\left(\frac{N}{2}\right)}\int_{0}^{R}rJ_{\frac{N}{2} - 1}^{2}\left(\frac{j_{\frac{N}{2} - 1,1}}{R}r\right)\hbox{  }dr = \frac{2\pi^{\frac{N}{2}}}{\Gamma\left(\frac{N}{2}\right)}\frac{R^{2}}{j^{2}_{\frac{N}{2} - 1, 1}}\int_{0}^{j_{\frac{N}{2} - 1,1}}s J^{2}_{\frac{N}{2} - 1}(s)\hbox{  }ds.
\end{align*}

Como queremos que $ \int_{\Omega}|\varphi_{1}(x)|^{2}\hbox{  }dx = 1 $, temos

$$ c^{2}\left( \frac{2\pi^{\frac{N}{2}}}{\Gamma\left(\frac{N}{2}\right)}\frac{R^{2}}{j^{2}_{\frac{N}{2} - 1, 1}}\int_{0}^{j_{\frac{N}{2} - 1,1}}s J^{2}_{\frac{N}{2} - 1}(s)\hbox{  }ds \right) = 1 $$

portanto

$$ c = \left( \dfrac{\Gamma\left(\frac{N}{2}\right)j^{2}_{\frac{N}{2} - 1,1}}{2\pi^{\frac{N}{2}}R^{2}\int_{0}^{j_{\frac{N}{2} - 1,1}}s J^{2}_{\frac{N}{2} - 1}(s)\hbox{  }ds} \right)^{\frac{1}{2}} $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn, jn_zeros, gamma
from scipy.integrate import quad

# Lista de valores de N para análise
N_values = [2, 3, 4, 5]

# Definição dos parâmetros
R = 1  # Raio da bola
k = 1000
num = k + 1

# Criando a figura
plt.figure(figsize=(8, 6))

# Cores para diferenciar as curvas
colors = ['b', 'g', 'm', 'c']

# Loop sobre os valores de N
for i, N in enumerate(N_values):
    n = N / 2 - 1  # Ordem da função de Bessel

    # Primeiro zero da função de Bessel de ordem n
    j_n1 = jn_zeros(int(np.floor(n)), 1)[0]

    # Função para a integral no denominador de c
    def integrand(s):
        return s * jn(n, s) ** 2

    # Cálculo da integral de 0 até j_n1
    integral_value, _ = quad(integrand, 0, j_n1)

    # Cálculo da constante c
    c = np.sqrt((gamma(N / 2) * j_n1**2) / (2 * np.pi**(N / 2) * R**2 * integral_value))

    # Função da solução exata com c correto
    def solucaoexata(r):
        return c * r**(1 - N / 2) * jn(n, j_n1 * r)

    # Evitar r = 0, começando de um valor muito pequeno
    t = np.linspace(1e-6, R, num)

    # Cálculo da solução exata para cada valor de r
    vr = solucaoexata(t)

    # Plotando a autofunção para cada N
    plt.plot(t, vr, color=colors[i], label=f'$N={N}$')

# Configuração do gráfico
plt.xlabel('$r$')
plt.ylabel('$v(r)$')
plt.title('Autofunções $v(r)$ para diferentes valores de $N$')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()


```

$\textbf{Entendendo PINNs}$

Antes de nos aprofundarmos como o PINNs funciona, é necessário entender o básico de redes neurais. Redes neurais tentão mimetizar o que sabemos sobre os neurônios humanos, por exemplo, em um neurônio humano os dendritos coletam sinais de entrada, nos quais são inseridos dentro do neurônio e pesados de acordo com sua importância, se esse valor atingir um certo limite o neurônio se ativa e emite um sinal através do axion para outros neurônios

**1 - Neurônios Individuais**

Um neurônio matemático é definido como uma função que leva $N$ valores reais como entrada e mapeia em um número real através da composição de uma transformação linear com ação de uma função não-linear. Isto é, dado um vetor de entrada $x = \{x_{1},\dots x_{N}\}$, o neurônio retorna $a(z)$, onde

$$ z = \sum_{i = 1}^{N}\omega_{i}x_{i} - b  $$
é uma soma com peso, e $a$ é chamada de função de ativação. Os coeficientes $W = \{\omega_{i}\}$ e $b$ são chamados de pesos e viés, respectivamente. De forma esquemática, a ação de um nerônio pode ser expressada da seguinte maneira:

\begin{align*}
&\mathbb{R}^{n}\longrightarrow\mathbb{R}\longrightarrow\mathbb{R}\\
&\{x_{n}\}\longrightarrow z \longrightarrow a(z).
\end{align*}

Note que $ \sum_{i = 1}^{N}\omega_{i}x_{i} > b$. A função de ativação tenta mimetizar a resposta de ativação de um neurônio real.

De forma geral, uma rede neural é uma série de camadas de neurônios

representação rede neural.png

Onde a informação flui da esquerda para direita. A primeira camada é a camada de entrada $x$, seguida pelas camadas ocultas. A camada de saída nos dá as previsões da rede neural, e é denotada por $y_{NN}(x,\theta)$ onde $\theta = [W^{k},b^{k}]$ para $0\le k \le l$, com $l$ sendo o número de camadas. Portanto, $\theta$ representa o conjunto dos pesos e viéses de todas as camadas ocultas em um particular passo de treinamento e depende do estado específico da rede durante o treinamento.

Todos os parâmetros podem ser manualmente trocados na rede neural, como o número de camadas, neurônio por camadas, rotinas de treinamento, etc e são chamados de hiperparâmetros. O problema é que não existe uma regra clara de como otimizar esses hiperparâmetros, então devemos fazer manualmente, possivelmente usando regras heurísticas.

**2- Função de Perda**

No uso geral de redes neurais, precisamos acessar os "dados verdadeiros", isto é, varias instâncias de pares $(x,y_{\mathrm{true}})$. Isso é referido na literatura como aprendizado supervisionado.

Como já foi dito, esperasse que a rede neural aprenda com a transformação entre as entradas e saídas, por tanto a "distância" entre as previsões e os valores reais devem tender a zero.

Essa distância entre valores reais e previsões em uma rede neural é chamada de função de perda $L$. A função de peso depende dos pesos e viéses de todos os neurônios na rede. Existem infinitas escolhas para $L$ porém a mais símples é o erro médio quadrático (MSE) entre as previsões $y_{NN}(x)$ e os valores reais $y_{\mathrm{true}}(x)$ de todos os dados de treino:

$$ MSE = \frac{1}{n}\sum_{i}|y_{\mathrm{true}}(x_{i}) - y_{NN}(x_{i})|^{2}, $$

onde $n$ é o número de data poins usado no cálculo da função.

Observe que como os pesos e viéses são inicializados com valores aleatórios, as previsões iniciais de $y_{NN}$ não são relacionadas a $y_{\mathrm{true}}$, e o valor inicial da função de perda é alto. Portanto o processo de aprendizado minimiza a função de perda modificando gradualmente os pesos e viéses.

Normalmente a minimização da função de perda é baseado no método do gradiente descendente, isto é, enquanto $\nabla L = \left( \frac{\partial L}{\partial\omega_{i}},\frac{\partial L}{\partial b_{i}} \right)$ aponta para o máximo da função, a busca pelo mínimo global é feita se movendo na direção oposta do gradiente. Em cada passo do gradiente descendente, atualizamo os valores para cada peso $\omega_{i}$ e cada viés $b_{i}$ de acordo com as equações

$$ \omega_{i}^{N+1} = \omega_{i}^{N} - \eta\frac{\partial L}{\partial b_{i}}\quad b_{i}^{N+1} = b_{i}^{N} - \eta\frac{\partial L}{\partial b_{i}} $$

onde o parâmetro $\eta$ é conhecido como taxa de aprendizado.

O gradiente descendente nos permite minimizar a função de perda mas ele não nos diz como calcular esses gradientes. Poderiamos calcular as derivadas parciais utilizando diferenças finitas, porém esse processo requer um número muito grande de operações e faria com que o processo de aprendizagem fosse extremamente lento e ineficiente.

Portanto para o caso das redes neurais, um dos métodos mais usados para calcular os gradientes é a retropropagação, que utiliza a regra da cadeia para computar os gradientes de forma eficiente.

**3 - Definição de PINNs**

As Physics-Informed Neural Networks (PINNs) são uma classe de redes neurais projetadas para desenvolver Equações Diferenciais (parciais ou ordnárias) incorporando diretamente as leis físicas na função de perda da rede. A idéia central é usar redes neurais para aproximar a solução de uma Equação Diferencial, enquanto as condições de contorno e iniciais são impostas como penalizações na função de perda.

A construção de uma PINN envolve as seguintes componentes:

 **3.1 - Rede Neural**

Uma rede neural $u_{\theta}$ é utilizada para aproximar a solução $u(x)$ de uma EDP, onde $x\in\Omega\subset\mathbb{R}^{N}$ é a variável de entrada (podendo ser as coordenadas espaciais ou temporais da equação) e $\theta$ é o parâmetro da rede (isto é, peso e viés)

**3.2 - Equação Diferencial**

Considere uma EDP geral da forma:

$$ \mathcal{F}(u,\nabla u,\nabla^{2}u,\dots,) = 0,\quad x\in\Omega$$

onde $\mathcal{F}$ é um operador diferencial (no nosso caso $\mathcal{F} = \Delta u$).

A equação diferencial é incorporada na função de perda como

\begin{equation}
\mathcal{L}_{res}(\theta) = \frac{1}{N_{r}}\sum_{i = 1}^{N_{r}}|\mathcal{F}(u_{\theta}(x_{i}),\nabla u_{\theta}(x_{i}),\nabla^{2}u_{\theta}(x_{i}),\dots)|^{2} \end{equation}

onde $\{x_{i}\}_{i = 1}^{N_{r}}$ são pontos amostrados no domínio $\Omega$.

**3.3 - Condições de Contorno/Iniciais**

Condições de contorno (ou iniciais) são impostas como penalizações adicionais, por exemplo para condição de Dirchlet $u = g$ em $\partial\Omega$, a função de perda é

\begin{equation}
\mathcal{L}_{bc}(\theta) = \frac{1}{N_{b}}\sum_{j = 1}^{N_{b}}|u_{\theta}(x_{j}) - g(x_{j})|^{2},
\end{equation}

onde $\{x_{j}\}^{N_{b}}_{j = 1}$ são os pontos amostrados na fronteira de $\partial\Omega$.

**3.4 - Função de Perda Total**

A função total de perda é uma combinação das outras funções de perda, isto é

$$ \mathcal{L}(\theta) = \mathcal{L}_{res}(\theta) + \mathcal{L}_{bc}(\theta). $$

**3.5 - Treinamento**

O treinamento consiste em minimizar a função de perda $\mathcal{L}(\theta)$ em relação aos parâmetros $\theta$ da rede neural:

$$ \theta^{*} = \arg\min_{\theta}\mathcal{L}(\theta). $$

Faremos isso usando o método do gradiente descendente.

**4 - Porquê usar PINNs?**

Existem muitas vantagens em utilizar PINNs para lidar com equações diferenciais como:

$\bullet$ Não necessita discretizar o domínio, o que é bastante útil quando se lida com geometrias complexas ou altas dimensões.

$\bullet$ Flexibilidade, pois pode ser aplicado em quase todo tipo de equações direfenciais.

$\bullet$ Facilidade de incorporação de dados, que podem ser facilmente utilizados nas funções de perda.

$\bullet$ As soluções são suaves (isto é diferenciáveis), o que é bastante útil quando se deseja fazer aplicações dessas equações.

**5 - Dificuldades das PINNs**

As PINNs no entanto não são perfeitas, elas possui um custo computacional considerável (principalmente quando se envolve altas dimensões), requer um cuidado na escolha de hiperparâmetros (como a arquitetura da rede, taxa de aprendizado e número de pontos amostrados) e possui problemas com convergência de soluções.


$\textbf{Formulando o Problema em PINNs}$

Voltando ao problema original, recordemos o problema de autovalor

\begin{align}
 \begin{cases}
\Delta u = \lambda u,\hbox{  }u\in\Omega\subset\mathbb{R}^{N}\\
\hbox{  }\hbox{  }\hbox{  }u \equiv 0,\quad u\in\partial \Omega
 \end{cases}
\end{align}

onde $\Omega$ é uma bola de raio $R$ e $\partial\Omega$ sua fronteira.

Para o nosso problema iremos precisar de três funções de perda, a da equação diferencial, da condição de contorno e da normalização da autofunção. Isto é,

$$ \mathcal{L}(\theta,\lambda) = \mathcal{L}_{res}(\theta,\mu) + \mathcal{L}_{bc}(\theta) + \mathcal{L}_{norm}(\theta), $$

onde

$$ \mathcal{L}_{res}(\theta,\lambda) = \frac{1}{N_{r}}\sum_{i = 1}^{N_{r}}|\Delta u_{\theta}(x_{i}) - \lambda u_{\theta}(x_{i})|^{2} $$

é a função de perda da equação diferencial,

$$ \mathcal{L}_{bc}(\theta) = \frac{1}{N_{b}}\sum_{ j = 1}^{N_{b}}|u_{\theta}(x_{j})|^{2}$$

é a função de perda da condição de contorno, e

$$ \mathcal{L}_{\mathrm{norm}}(\theta) = \left| \frac{1}{N_{r}}\sum_{i = 1}^{N_{r}}u_{\theta}(x_{i})^{2} - 1 \right|^{2} $$

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from scipy.special import jn, gamma
from scipy.integrate import quad

# Parâmetros do problema
N = 1  # Dimensão do espaço (N = 2 para 2D)
R = 1.0  # Raio da bola
n = 1  # Ordem da função de Bessel
j_n1 = 3.8317  # Primeiro zero da função de Bessel J_n (para n = 1)

# Solução exata
def solucao_exata(r):
    # Constante de normalização
    def integrando(r):
        return (r**(1 - N / 2) * jn(n, j_n1 * r))**2 * r**(N - 1)
    integral_value, _ = quad(integrando, 0, R)
    c = np.sqrt((gamma(N / 2) * j_n1**2) / (2 * np.pi**(N / 2) * R**2 * integral_value))
    return c * r**(1 - N / 2) * jn(n, j_n1 * r)

# Rede neural para a solução via PINNs
def neural_net(x, layers, activation):
    for i, size in enumerate(layers):
        x = tf.keras.layers.Dense(size, activation=activation)(x)
    return x

# Função de perda para PINNs
def loss_fn(u_net, lambda_var, x_r, x_b):
    # Perda da equação diferencial
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(x_r)
        u = u_net(x_r)
        u_r = tape.gradient(u, x_r)
        u_rr = tape.gradient(u_r, x_r)
    loss_res = tf.reduce_mean(tf.square(u_rr + (N - 1) / x_r * u_r - lambda_var * u))

    # Perda da condição de contorno
    u_b = u_net(x_b)
    loss_bc = tf.reduce_mean(tf.square(u_b))

    # Perda de normalização
    loss_norm = tf.square(tf.reduce_mean(u**2) - 1)

    return loss_res + loss_bc + loss_norm

# Treinamento da PINN
def train(u_net, lambda_var, x_r, x_b, lr=0.001, epochs=1000):
    optimizer = tf.optimizers.Adam(learning_rate=lr)
    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            loss = loss_fn(u_net, lambda_var, x_r, x_b)
        grads = tape.gradient(loss, u_net.trainable_variables + [lambda_var])
        optimizer.apply_gradients(zip(grads, u_net.trainable_variables + [lambda_var]))
        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss.numpy()}")

# Amostragem de pontos
N_r = 100  # Número de pontos no interior
N_b = 50   # Número de pontos na fronteira
x_r = tf.convert_to_tensor(np.linspace(0.01, R, N_r)[:, np.newaxis], dtype = tf.float32)  # Pontos no interior (evitar r = 0)
x_b = tf.convert_to_tensor(np.array([[R]]), dtype = tf.float32)  # Pontos na fronteira

# Construção da rede neural
layers = [1, 20, 20, 1]  # Arquitetura da rede
activation = 'tanh'      # Função de ativação
inputs = tf.keras.Input(shape=(1,))
outputs = neural_net(inputs, layers, activation)
u_net = tf.keras.Model(inputs, outputs)

# Inicialização do autovalor
lambda_var = tf.Variable(1.0, dtype=tf.float32)

# Treinamento
train(u_net, lambda_var, x_r, x_b, lr=0.001, epochs=2000)

# Solução exata
r_values = np.linspace(0, R, 100)
u_exata = solucao_exata(r_values)

# Solução via PINNs
u_pinn = u_net(r_values[:, np.newaxis]).numpy().flatten()

# Plotagem das soluções
plt.figure(figsize=(8, 6))
plt.plot(r_values, u_exata, label="Solução Exata", linewidth=2)
plt.plot(r_values, u_pinn, '--', label="Solução PINN", linewidth=2)
plt.xlabel("r")
plt.ylabel("u(r)")
plt.title("Comparação entre Solução Exata e PINN")
plt.legend()
plt.grid()
plt.show()
```
